```python
import pandas as pd
import seaborn as sns
import matplotlib as plt
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('customer_data.csv')
```

```python
sns.heatmap(data)
```

```python
X = data.drop(["purchased"],axis="columns")
X
```

```python
Y = data['purchased']
Y
```

```python
X_train,X_test,Y_train,Y_real=train_test_split(X,Y,train_size = 0.8)
```

# Logistic Regression

```python
model_logistic = LogisticRegression(solver='liblinear')
```

```python
model_logistic.fit(X_train,Y_train)
```

```python
y_predicted_logistic = model_logistic.predict(X_test)
```

```python
logistic_regression_accuracy = model_logistic.score(X_test,Y_real)
```

```python
metrics.accuracy_score(Y_real,y_predicted_logistic)
```

```python
# Precision = true positive / true positive + false positive
precision_lg = metrics.precision_score(Y_real,y_predicted_logistic)  
```

```python
# Recall = true positive / true positive + false negative
rec_lg = metrics.recall_score(Y_real, y_predicted_logistic)
rec_lg
```

```python
cm_lg = metrics.confusion_matrix(Y_real,y_predicted_logistic)
cm_lg
```

```python
sns.heatmap(cm_lg, annot=True)
```
